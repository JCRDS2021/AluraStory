```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

# **FATURAMENTO**

```python
# REALIZANDO A SOMA DO FATURAMENTO DE CADA LOJA

fat_loja = round(sum(loja['Preço']),2)
fat_loja2 = round(sum(loja2['Preço']),2)
fat_loja3 = round(sum(loja3['Preço']),2)
fat_loja4 = round(sum(loja4['Preço']),2)
```

```python
# PRINTANDO O FATURAMENTO DE CADA LOJA

print(f'Faturamento Loja 1: {fat_loja}')
print(f'Faturamento Loja 2: {fat_loja2}')
print(f'Faturamento Loja 3: {fat_loja3}')
print(f'Faturamento Loja 4: {fat_loja4}')
```

```python
# REALIZANDO A SOMA DAS VENDAS POR CATEGORIA DE PRODUTO

vendas_cat_loja = loja.groupby('Categoria do Produto')['Preço'].sum()
vendas_cat_loja2 = loja2.groupby('Categoria do Produto')['Preço'].sum()
vendas_cat_loja3 = loja3.groupby('Categoria do Produto')['Preço'].sum()
vendas_cat_loja4 = loja4.groupby('Categoria do Produto')['Preço'].sum()
```

# **MÉDIA DE AVALIAÇÃO**

```python
# REALIZANDO A MÉDIA DE AVALIAÇÃO DE CADA LOJA

media_aval_loja = round(loja['Avaliação da compra'].mean(),2)
media_aval_loja2 = round(loja2['Avaliação da compra'].mean(),2)
media_aval_loja3 = round(loja3['Avaliação da compra'].mean(),2)
media_aval_loja4 = round(loja4['Avaliação da compra'].mean(),2)
```

```python
# PRINTANDO A MÉDIA DE AVALIAÇÃO DE CADA LOJA

print(f'Avaliação da Loja: {media_aval_loja}')
print(f'Avaliação da Loja 2: {media_aval_loja2}')
print(f'Avaliação da Loja 3: {media_aval_loja3}')
print(f'Avaliação da Loja 4: {media_aval_loja4}')
```

# **PRODUTOS MAIS E MENOS VENDIDOS**

```python
# FAZENDO O AGRUPAMENTO DAS VENDAS POR PRODUTOS

venda_produto_loja = pd.DataFrame(loja.groupby('Produto')['Preço'].sum())
venda_produto_loja2 = pd.DataFrame(loja2.groupby('Produto')['Preço'].sum())
venda_produto_loja3 = pd.DataFrame(loja3.groupby('Produto')['Preço'].sum())
venda_produto_loja4 = pd.DataFrame(loja4.groupby('Produto')['Preço'].sum())
```

```python
# TOP 10 MAIS VENDIDOS DE CADA LOJA

top10_loja = pd.DataFrame(venda_produto_loja.sort_values('Preço', ascending=False).head(10))
top10_loja2 = pd.DataFrame(venda_produto_loja2.sort_values('Preço', ascending=False).head(10))
top10_loja3 = pd.DataFrame(venda_produto_loja3.sort_values('Preço', ascending=False).head(10))
top10_loja4 = pd.DataFrame(venda_produto_loja4.sort_values('Preço', ascending=False).head(10))

# 10 MENOS VENDIDOS DE CADA LOJA

menos10_loja = venda_produto_loja.sort_values('Preço', ascending=False).tail(10)
menos10_loja2 = venda_produto_loja2.sort_values('Preço', ascending=False).tail(10)
menos10_loja3 = venda_produto_loja3.sort_values('Preço', ascending=False).tail(10)
menos10_loja4 = venda_produto_loja4.sort_values('Preço', ascending=False).tail(10)
```

```python
# PRINTANDO OS 10 MAIS VENDIDOS DE CADA LOJA E OS 10 MENOS VENDIDOS DE CADA LOJA

print('TOP 10 MAIS VENDIDOS DE CADA LOJA')
print(top10_loja)
print(top10_loja2)
print(top10_loja3)
print(top10_loja4)

print()

print('TOP 10 MENOS VENDIDOS DE CADA LOJA')
print(menos10_loja)
print(menos10_loja2)
print(menos10_loja3)
print(menos10_loja4)
```

# **FRETE MÉDIO POR LOJA**

```python
# CALCULANDO O FRETE MÉDIO POR LOJA

frete_medio_loja = round(loja['Frete'].mean(),2)
frete_medio_loja2 = round(loja2['Frete'].mean(),2)
frete_medio_loja3 = round(loja3['Frete'].mean(),2)
frete_medio_loja4 = round(loja4['Frete'].mean(),2)
```

```python
# PRINTANDO OS RESULTADOS DOS FRETES

print('FRETES MÉDIOS POR LOJA')
print(f'Loja 1: {frete_medio_loja}')
print(f'Loja 2: {frete_medio_loja2}')
print(f'Loja 3: {frete_medio_loja3}')
print(f'Loja 4: {frete_medio_loja4}')
```

# **VISUALIZAÇÕES**

```python
dados_estrategicos = pd.DataFrame({'LOJAS': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
                                   'FATURAMENTO': [fat_loja, fat_loja2, fat_loja3, fat_loja4],
                                   'MÉDIA DE AVALIAÇÃO': [media_aval_loja, media_aval_loja2, media_aval_loja3, media_aval_loja4],
                                   'FRETE MÉDIA POR LOJA': [frete_medio_loja, frete_medio_loja2, frete_medio_loja3, frete_medio_loja4]})
```

```python
import matplotlib.pyplot as plt
```

```python
# GRÁFICO DE FATURAMENTO TOTAL DE TODAS AS LOJAS (BARRAS)

# Dados para o gráfico
lojas = dados_estrategicos['LOJAS']
faturamento = dados_estrategicos['FATURAMENTO']

# Criando o gráfico de barras
plt.bar(lojas, faturamento, color='green')

# Personalizando o gráfico
plt.title('Faturamento total de cada loja')  # Título
plt.xlabel('Lojas')  # Nome do eixo X
plt.ylabel('Faturamento')  # Nome do eixo Y

# Exibindo o gráfico
plt.show()
```

```python
# GRÁFICO DE MÉDIA DE AVALIAÇÃO DE TODAS AS LOJAS (LINHAS)

lojas = dados_estrategicos['LOJAS']
media_avaliacao = dados_estrategicos['MÉDIA DE AVALIAÇÃO']

plt.plot(lojas, media_avaliacao, marker='o')
plt.title("Média de Avaliação")
plt.xlabel("Lojas")
plt.ylabel("Média de Avaliação")
plt.show()
```

```python
# GRÁFICO DO FRETE MÉDIO POR LOJA (VALORES ARREDONDADOS) - PIZZA

# Dados do gráfico
labels = dados_estrategicos['LOJAS']
valores = [frete_medio_loja, frete_medio_loja2, frete_medio_loja3, frete_medio_loja4]
# Criar o gráfico de pizza
plt.pie(valores, labels=labels, autopct=lambda pct: f'{int(round(pct/100.*sum(valores)))}')
plt.title("Frete Médio por Loja")
# Exibir o gráfico
plt.show()
```

```python
# GRÁFICO DE LINHA REPRESENTANDO OS 10 PRODUTOS MAIS VENDIDOS DE CADA LOJA
# PARA VERIFICAR OS RESULTADOS DE CADA LOJA, BASTA ADICIONAR O NÚMERO DA LOJA NA SEGUNDA LINHA
# EM top10_loja(número da loja). Obs.: top10_loja representa a primeira loja.

plt.figure(figsize=(20, 8))
plt.plot(top10_loja4, marker='o')
plt.title("Produtos mais Vendidos da Loja 1")
plt.xlabel("Produtos")
plt.ylabel("Total Vendido")
plt.show()
```

```python
# GRÁFICO DE LINHA REPRESENTANDO OS 10 PRODUTOS MENOS VENDIDOS DE CADA LOJA
# PARA VERIFICAR OS RESULTADOS DE CADA LOJA, BASTA ADICIONAR O NÚMERO DA LOJA NA SEGUNDA LINHA
# EM menos10_loja(número da loja). Obs.: menos10_loja representa a primeira loja.

plt.figure(figsize=(20, 8))
plt.plot(menos10_loja4, marker='o')
plt.title("Produtos menos Vendidos da Loja 1")
plt.xlabel("Produtos")
plt.ylabel("Total Vendido")
plt.show()
```

```python
# GRÁFICO DE LINHA REPRESENTANDO AS CATEGORIAS MAIS VENDIDAS DE CADA LOJA
# PARA VERIFICAR OS RESULTADOS DE CADA LOJA, BASTA ADICIONAR O NÚMERO DA LOJA NA SEGUNDA LINHA
# EM vendas_cat_loja(número da loja). Obs.: vendas_cat_loja representa a primeira loja.

plt.figure(figsize=(20, 8))
plt.plot(vendas_cat_loja2, marker='o')
plt.title("Produtos mais Vendidos da Loja 4")
plt.xlabel("Produtos")
plt.ylabel("Total Vendido")
plt.show()
```
