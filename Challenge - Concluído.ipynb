```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

# **FATURAMENTO**

```python
# REALIZANDO A SOMA DO FATURAMENTO DE CADA LOJA

fat_loja = round(sum(loja['Pre√ßo']),2)
fat_loja2 = round(sum(loja2['Pre√ßo']),2)
fat_loja3 = round(sum(loja3['Pre√ßo']),2)
fat_loja4 = round(sum(loja4['Pre√ßo']),2)
```

```python
# PRINTANDO O FATURAMENTO DE CADA LOJA

print(f'Faturamento Loja 1: {fat_loja}')
print(f'Faturamento Loja 2: {fat_loja2}')
print(f'Faturamento Loja 3: {fat_loja3}')
print(f'Faturamento Loja 4: {fat_loja4}')
```

```python
# REALIZANDO A SOMA DAS VENDAS POR CATEGORIA DE PRODUTO

vendas_cat_loja = loja.groupby('Categoria do Produto')['Pre√ßo'].sum()
vendas_cat_loja2 = loja2.groupby('Categoria do Produto')['Pre√ßo'].sum()
vendas_cat_loja3 = loja3.groupby('Categoria do Produto')['Pre√ßo'].sum()
vendas_cat_loja4 = loja4.groupby('Categoria do Produto')['Pre√ßo'].sum()
```

# **M√âDIA DE AVALIA√á√ÉO**

```python
# REALIZANDO A M√âDIA DE AVALIA√á√ÉO DE CADA LOJA

media_aval_loja = round(loja['Avalia√ß√£o da compra'].mean(),2)
media_aval_loja2 = round(loja2['Avalia√ß√£o da compra'].mean(),2)
media_aval_loja3 = round(loja3['Avalia√ß√£o da compra'].mean(),2)
media_aval_loja4 = round(loja4['Avalia√ß√£o da compra'].mean(),2)
```

```python
# PRINTANDO A M√âDIA DE AVALIA√á√ÉO DE CADA LOJA

print(f'Avalia√ß√£o da Loja: {media_aval_loja}')
print(f'Avalia√ß√£o da Loja 2: {media_aval_loja2}')
print(f'Avalia√ß√£o da Loja 3: {media_aval_loja3}')
print(f'Avalia√ß√£o da Loja 4: {media_aval_loja4}')
```

# **PRODUTOS MAIS E MENOS VENDIDOS**

```python
# FAZENDO O AGRUPAMENTO DAS VENDAS POR PRODUTOS

venda_produto_loja = pd.DataFrame(loja.groupby('Produto')['Pre√ßo'].sum())
venda_produto_loja2 = pd.DataFrame(loja2.groupby('Produto')['Pre√ßo'].sum())
venda_produto_loja3 = pd.DataFrame(loja3.groupby('Produto')['Pre√ßo'].sum())
venda_produto_loja4 = pd.DataFrame(loja4.groupby('Produto')['Pre√ßo'].sum())
```

```python
# TOP 10 MAIS VENDIDOS DE CADA LOJA

top10_loja = pd.DataFrame(venda_produto_loja.sort_values('Pre√ßo', ascending=False).head(10))
top10_loja2 = pd.DataFrame(venda_produto_loja2.sort_values('Pre√ßo', ascending=False).head(10))
top10_loja3 = pd.DataFrame(venda_produto_loja3.sort_values('Pre√ßo', ascending=False).head(10))
top10_loja4 = pd.DataFrame(venda_produto_loja4.sort_values('Pre√ßo', ascending=False).head(10))

# 10 MENOS VENDIDOS DE CADA LOJA

menos10_loja = venda_produto_loja.sort_values('Pre√ßo', ascending=False).tail(10)
menos10_loja2 = venda_produto_loja2.sort_values('Pre√ßo', ascending=False).tail(10)
menos10_loja3 = venda_produto_loja3.sort_values('Pre√ßo', ascending=False).tail(10)
menos10_loja4 = venda_produto_loja4.sort_values('Pre√ßo', ascending=False).tail(10)
```

```python
# PRINTANDO OS 10 MAIS VENDIDOS DE CADA LOJA E OS 10 MENOS VENDIDOS DE CADA LOJA

print('TOP 10 MAIS VENDIDOS DE CADA LOJA')
print(top10_loja)
print(top10_loja2)
print(top10_loja3)
print(top10_loja4)

print()

print('TOP 10 MENOS VENDIDOS DE CADA LOJA')
print(menos10_loja)
print(menos10_loja2)
print(menos10_loja3)
print(menos10_loja4)
```

# **FRETE M√âDIO POR LOJA**

```python
# CALCULANDO O FRETE M√âDIO POR LOJA

frete_medio_loja = round(loja['Frete'].mean(),2)
frete_medio_loja2 = round(loja2['Frete'].mean(),2)
frete_medio_loja3 = round(loja3['Frete'].mean(),2)
frete_medio_loja4 = round(loja4['Frete'].mean(),2)
```

```python
# PRINTANDO OS RESULTADOS DOS FRETES

print('FRETES M√âDIOS POR LOJA')
print(f'Loja 1: {frete_medio_loja}')
print(f'Loja 2: {frete_medio_loja2}')
print(f'Loja 3: {frete_medio_loja3}')
print(f'Loja 4: {frete_medio_loja4}')
```

# **VISUALIZA√á√ïES**

```python
dados_estrategicos = pd.DataFrame({'LOJAS': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
                                   'FATURAMENTO': [fat_loja, fat_loja2, fat_loja3, fat_loja4],
                                   'M√âDIA DE AVALIA√á√ÉO': [media_aval_loja, media_aval_loja2, media_aval_loja3, media_aval_loja4],
                                   'FRETE M√âDIA POR LOJA': [frete_medio_loja, frete_medio_loja2, frete_medio_loja3, frete_medio_loja4]})
```

```python
import matplotlib.pyplot as plt
```

```python
# GR√ÅFICO DE FATURAMENTO TOTAL DE TODAS AS LOJAS (BARRAS)

# Dados para o gr√°fico
lojas = dados_estrategicos['LOJAS']
faturamento = dados_estrategicos['FATURAMENTO']

# Criando o gr√°fico de barras
plt.bar(lojas, faturamento, color='green')

# Personalizando o gr√°fico
plt.title('Faturamento total de cada loja')  # T√≠tulo
plt.xlabel('Lojas')  # Nome do eixo X
plt.ylabel('Faturamento')  # Nome do eixo Y

# Exibindo o gr√°fico
plt.show()
```

```python
# GR√ÅFICO DE M√âDIA DE AVALIA√á√ÉO DE TODAS AS LOJAS (LINHAS)

lojas = dados_estrategicos['LOJAS']
media_avaliacao = dados_estrategicos['M√âDIA DE AVALIA√á√ÉO']

plt.plot(lojas, media_avaliacao, marker='o')
plt.title("M√©dia de Avalia√ß√£o")
plt.xlabel("Lojas")
plt.ylabel("M√©dia de Avalia√ß√£o")
plt.show()
```

```python
# GR√ÅFICO DO FRETE M√âDIO POR LOJA (VALORES ARREDONDADOS) - PIZZA

# Dados do gr√°fico
labels = dados_estrategicos['LOJAS']
valores = [frete_medio_loja, frete_medio_loja2, frete_medio_loja3, frete_medio_loja4]
# Criar o gr√°fico de pizza
plt.pie(valores, labels=labels, autopct=lambda pct: f'{int(round(pct/100.*sum(valores)))}')
plt.title("Frete M√©dio por Loja")
# Exibir o gr√°fico
plt.show()
```

```python
# GR√ÅFICO DE LINHA REPRESENTANDO OS 10 PRODUTOS MAIS VENDIDOS DE CADA LOJA
# PARA VERIFICAR OS RESULTADOS DE CADA LOJA, BASTA ADICIONAR O N√öMERO DA LOJA NA SEGUNDA LINHA
# EM top10_loja(n√∫mero da loja). Obs.: top10_loja representa a primeira loja.

plt.figure(figsize=(20, 8))
plt.plot(top10_loja4, marker='o')
plt.title("Produtos mais Vendidos da Loja 1")
plt.xlabel("Produtos")
plt.ylabel("Total Vendido")
plt.show()
```

```python
# GR√ÅFICO DE LINHA REPRESENTANDO OS 10 PRODUTOS MENOS VENDIDOS DE CADA LOJA
# PARA VERIFICAR OS RESULTADOS DE CADA LOJA, BASTA ADICIONAR O N√öMERO DA LOJA NA SEGUNDA LINHA
# EM menos10_loja(n√∫mero da loja). Obs.: menos10_loja representa a primeira loja.

plt.figure(figsize=(20, 8))
plt.plot(menos10_loja4, marker='o')
plt.title("Produtos menos Vendidos da Loja 1")
plt.xlabel("Produtos")
plt.ylabel("Total Vendido")
plt.show()
```

```python
# GR√ÅFICO DE LINHA REPRESENTANDO AS CATEGORIAS MAIS VENDIDAS DE CADA LOJA
# PARA VERIFICAR OS RESULTADOS DE CADA LOJA, BASTA ADICIONAR O N√öMERO DA LOJA NA SEGUNDA LINHA
# EM vendas_cat_loja(n√∫mero da loja). Obs.: vendas_cat_loja representa a primeira loja.

plt.figure(figsize=(20, 8))
plt.plot(vendas_cat_loja2, marker='o')
plt.title("Produtos mais Vendidos da Loja 4")
plt.xlabel("Produtos")
plt.ylabel("Total Vendido")
plt.show()
```
